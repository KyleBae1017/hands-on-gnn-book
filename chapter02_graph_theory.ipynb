### Weighted Graphs

```python
WG = nx.Graph()
WG.add_edges_from([('A', 'B', {"weight": 10}), ('A', 'C', {"weight": 20}), ('B', 'D', {"weight": 30}), ('B', 'E', {"weight": 40}), ('C', 'F', {"weight": 50}), ('C', 'G', {"weight": 60})])
labels = nx.get_edge_attributes(WG, "weight")

plt.axis('off')
nx.draw_networkx(WG,
                 pos=nx.spring_layout(G, seed=0),
                 node_size=600,
                 cmap='coolwarm',
                 font_size=14,
                 font_color='white'
                 )
nx.draw_networkx_edge_labels(WG, pos=nx.spring_layout(G, seed=0), edge_labels=labels)
```

### Connectivity

```python
G1 = nx.Graph()
G1.add_edges_from([(1, 2), (2, 3), (3, 1), (4, 5)])
plt.axis('off')
nx.draw_networkx(G1,
                 node_size=600,
                 font_size=14,
                 font_color='white'
                 )
print(f"Is graph 1 connected? {nx.is_connected(G1)}")

G2 = nx.Graph()
G2.add_edges_from([(1, 2), (2, 3), (3, 1), (1, 4)])
nx.draw_networkx(G2,
                 node_size=600,
                 node_color='r',
                 font_size=14,
                 font_color='white'
                 )
print(f"Is graph 2 connected? {nx.is_connected(G2)}")
```

```python
# strongly connected components (SCC) in directed graphs
G = nx.cycle_graph(4, create_using=nx.DiGraph())
nx.add_cycle(G, [10, 11, 12])

plt.axis('off')
nx.draw_networkx(G,
                 node_size=600,
                 font_size=14,
                 font_color='white'
                 )

for c in nx.strongly_connected_components(G):
  print(c)
```

### Degrees and Centrality

```python
G = nx.Graph()
G.add_edges_from([('A', 'B'), ('A', 'C'), ('B', 'D'), ('B', 'E'), ('C', 'F'), ('C', 'G')])
print(f"deg(A) = {G.degree['A']}")

DG = nx.DiGraph()
DG.add_edges_from([('A', 'B'), ('A', 'C'), ('B', 'D'), ('B', 'E'), ('C', 'F'), ('C', 'G')])
print(f"deg^-(A) = {DG.in_degree['A']}")
print(f"deg^+(A) = {DG.out_degree['A']}")

plt.axis('off')
plt.figure(1)
nx.draw_networkx(G,
                 pos=nx.circular_layout(G),
                 node_size=600,
                 font_size=14,
                 font_color='white'
                 )

plt.figure(2)
plt.axis('off')
nx.draw_networkx(DG,
                 pos=nx.circular_layout(G),
                 node_size=600,
                 font_size=14,
                 node_color='r',
                 font_color='white'
                 )
plt.show()
```

### Graph Search Algorithms : BFS, DFS

스크린샷 2023-07-12 오후 1.36.14.png
*pseudocode from AIMA*

```python
# BFS (fringe : queue)

G = nx.Graph()
G.add_edges_from([('A', 'B'), ('A', 'C'), ('B', 'D'), ('B', 'E'), ('C', 'F'), ('C', 'G')])
plt.axis('off')
nx.draw_networkx(G,
                 pos=nx.circular_layout(G),
                 node_size=600,
                 font_size=14,
                 font_color='white'
                 )

def bfs(graph, node):
    visited, queue = [node], [node]

    while queue:
        node = queue.pop(0)
        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.append(neighbor)
                queue.append(neighbor)

    return visited

bfs(G, 'A')
```

```python
def bfs(graph, node): # pseudocode-like
    visited, queue = [], [node]

    while queue:
        node = queue.pop(0)
        if node not in visited:
          visited.append(node)
          for neighbor in graph[node]: queue.append(neighbor)

    return visited

bfs(G, 'A')
```

```python
visited = []

def dfs(visited, graph, node):
    if node not in visited:
        visited.append(node)
        for neighbor in graph[node]:
            visited = dfs(visited, graph, neighbor)

    return visited

dfs(visited, G, 'A')
```

```python
def dfs_iter(graph, node):

    visited, stack = [], [node]

    while stack:
        node = stack.pop()
        if node not in visited:
          visited.append(node)
          for neighbor in graph[node]: stack.append(neighbor)

    return visited

dfs_iter(G, 'A')
```
